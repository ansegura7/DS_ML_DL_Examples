```python
# Import libraries
import numpy as np
from timeit import default_timer as timer
import matplotlib.pyplot as plt
```

```python
# Limits
test_list = [10000, 100000, 1000000, 10000000, 100000000]
N = max(test_list)
max_value = 10000
```

## 1. Uniform Distribution

```python
# Generate random uniform integers
uni_list = np.random.randint(0, max_value, N)
```

```python
# Plot uniform distribution
plt.rcParams["figure.figsize"] = (8,5)
plt.hist(uni_list, bins='auto')
plt.title('Uniform Distribution', size=16)
plt.xlabel('x')
plt.ylabel('freq')
plt.show()
```

```python
# Measuring time
time_uni_list = []

for peak in test_list:
  temp_list = uni_list[:peak]

  start_time = timer()
  temp_list.sort()
  elapsed_time = (timer() - start_time)
  time_uni_list.append(elapsed_time)
```

## 2. Normal Distribution

$$ f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^2} $$



```python
# Generate random normal numbers
mu, sigma = 0, 1
nor_list = np.random.normal(mu, sigma, size=N)
```

```python
# Scaling numbers
x_min, x_max = min(nor_list), max(nor_list)
nor_list = (nor_list - x_min) / (x_max - x_min)

# Round numbers to int
nor_list = [round(max_value * v) for v in nor_list]
```

```python
# Plot normal distribution
plt.rcParams["figure.figsize"] = (8,5)
plt.hist(nor_list, bins='auto')
plt.title('Normal Distribution', size=16)
plt.xlabel('x')
plt.ylabel('freq')
plt.show()
```

```python
# Measuring time
time_nor_list = []

for peak in test_list:
  temp_list = nor_list[:peak]

  start_time = timer()
  temp_list.sort()
  elapsed_time = (timer() - start_time)
  time_nor_list.append(elapsed_time)
```

## 3. Plot Results

```python
# Plot analysis results
plt.rcParams["figure.figsize"] = (10,5)
plt.plot(test_list, time_uni_list, linestyle='--', marker='o', label='Uniform')
plt.plot(test_list, time_nor_list, linestyle='--', marker='o', label='Normal')
plt.legend()
plt.title('Plot Analysis Results', size=16)
plt.xlabel('n')
plt.ylabel('time (s)')
plt.show()
```
