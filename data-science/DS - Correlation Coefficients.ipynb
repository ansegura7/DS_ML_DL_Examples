**Correlation** is the degree and direction in which two variables are related or associated. Therefore, **correlation coefficients** are used to measure the strength of the relationship between two variables.

A correlation can be positive, meaning that the 2 variables move in the same direction, or negative, meaning that they move in opposite directions. When the correlation is neutral (close to 0) it means that the 2 variables are unrelated.

Two of the most important correlation coefficients are:
- Pearson's correlation coefficient is a statistical measure of the strength of the linear relationship between two numerical variables. It is a very useful coefficient , but it is also very sensitive to outliers.
- Spearman correlation coefficient is a statistical measure of the strength of a monotonic relationship between two variables. Spearman correlation is less sensitive than the Pearson correlation to outliers.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Positive correlation

```python
# Parameters
w = 2.0
b = 1.0
xmin = 0.0
xmax = 10.0
noise = 4.0
n = 100

# Randomly generated problem data
x = xmin + np.random.rand(n)*(xmax - xmin)
y = w*x + b + np.random.randn(n)*noise
```

```python
# Calculate Pearson correlation
prho, pval = stats.pearsonr(x, y)
prho
```

```python
# Calculate Spearman correlation
srho, pval = stats.spearmanr(x, y)
srho
```

```python
# Scatter plot
def scatter_plot(x, y, title, prho, srho):
  plt.figure(figsize=(7, 7))
  plt.plot(x, y, 'o')
  plt.grid(True)
  plt.xlabel("x")
  plt.ylabel("y")
  plt.suptitle(title, fontsize=16)
  plt.title("Pearson {:.2f} and Spearman {:.2f}".format(prho, srho), fontsize=12)
  plt.show()
```

```python
# Plot results
scatter_plot(x, y, "Positive Correlation", prho, srho)
```

## Negative correlation

```python
# Parameters
w = -2.0
b = 1.0
xmin = 0.0
xmax = 10.0
noise = 4.0
n = 100

# Randomly generated problem data
x = xmin + np.random.rand(n)*(xmax - xmin)
y = w*x + b + np.random.randn(n)*noise
```

```python
# Calculate Pearson correlation
prho, pval = stats.pearsonr(x, y)
prho
```

```python
# Calculate Spearman correlation
srho, pval = stats.spearmanr(x, y)
srho
```

```python
# Plot results
scatter_plot(x, y, "Negative Correlation", prho, srho)
```

## Neutral correlation

```python
# Parameters
w = 1.0
b = 1.0
xmin = 0.0
xmax = 10.0
noise = 20.0
n = 100

# Randomly generated problem data
x = xmin + np.random.rand(n)*(xmax - xmin)
y = w*x + b + np.random.randn(n)*noise
```

```python
# Calculate Pearson correlation
prho, pval = stats.pearsonr(x, y)
prho
```

```python
# Calculate Spearman correlation
srho, pval = stats.spearmanr(x, y)
srho
```

```python
# Plot results
scatter_plot(x, y, "Neutral Correlation", prho, srho)
```

## Monotonic correlation

```python
# Parameters
mu = 0
sigma = 0.1
xmin = 0.0
xmax = 10.0
n = 100

# Randomly generated problem data
x = np.random.normal(mu, sigma, size=n)
x = (x - min(x)) * (xmax - xmin)
x.sort()
y = np.cumsum(x)
```

```python
# Calculate Pearson correlation
prho, pval = stats.pearsonr(x, y)
prho
```

```python
# Calculate Spearman correlation
srho, pval = stats.spearmanr(x, y)
srho
```

```python
# Plot results
scatter_plot(x, y, "Monotonic Correlation", prho, srho)
```
