# Neural Network that implements the XOR gate
### Solving XOR with a NN with a hidden layer

Created by Andres Segura-Tinoco  
Created on Mar 29, 2021

```python
# Importing Keras libraries
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
from tensorflow.keras import initializers
from keras.optimizers import Adam
from keras.utils.vis_utils import plot_model
import matplotlib.pyplot as plt
```

## 1. Create NN Model

```python
# Layer weight initializers
initializer = initializers.RandomNormal(mean=0.0, stddev=0.5, seed=None)
values = initializer(shape=(2, 2))
values
```

```python
# Create model
model = Sequential(name="XOR_MLP")
model.add(Dense(units=4, input_dim=2, use_bias=True, activation='relu', kernel_initializer=initializer, name='HL'))
model.add(Dense(units=1, use_bias=True, activation='sigmoid', kernel_initializer=initializer, name='Output'))

# Adam optimizer with learning rate of 0.02
model.compile(optimizer=Adam(lr=0.02), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

## 2. Train model with Binary data

```python
# Define binary encoding
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
target_data = np.array([[0],[1],[1],[0]], "float32")
```

```python
print("Fit model on training data")
results = model.fit(training_data, target_data, epochs=100, verbose=False)
```

```python
print(model.predict(training_data).round())
```

```python
# Summarize history for accuracy
plt.plot(results.history['accuracy'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.show()
```

## 3. Train model with Bipolar input data

```python
# Define bipolar encoding
training_data = np.array([[-1,-1],[-1,1],[1,-1],[1,1]], "float32")
target_data = np.array([[0],[1],[1],[0]], "float32")
```

```python
print("Fit model on training data")
results = model.fit(training_data, target_data, epochs=10, verbose=False)
```

```python
print(model.predict(training_data).round())
```

```python
# Summarize history for accuracy
plt.plot(results.history['accuracy'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.show()
```
